```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

# Importing machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Loading the dataset
heart_data = pd.read_csv('heart_disease_data.csv')

# Displaying the first 5 rows of the dataset
heart_data.head()

# Checking the shape (rows, columns) of the dataset
heart_data.shape

# Checking for missing values
heart_data.isnull().sum()

# Getting a statistical summary of the dataset
heart_data.describe()

# Checking the distribution of the target variable
heart_data['target'].value_counts()
# 1 - defective heart (disease present)
# 0 - healthy heart (no disease)

# Splitting the features and target
X = heart_data.drop(columns='target', axis=1)  # Feature variables
Y = heart_data['target']                       # Target variable

# Printing features and target
print(X)
print(Y)

# Splitting the data into training and testing sets
# 80% training data and 20% testing data, stratified based on target variable
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)

# Printing the shapes of the original, training, and testing datasets
print(X.shape, X_train.shape, X_test.shape)

# List of models to compare
models = [
    LogisticRegression(max_iter=1000),   # Logistic Regression model
    SVC(kernel='linear'),                # Support Vector Classifier with linear kernel
    KNeighborsClassifier(),              # K-Nearest Neighbors Classifier
    RandomForestClassifier()             # Random Forest Classifier
]

# Function to train and compare the performance of each model
def compare_models():
    for model in models:
        model.fit(X_train, Y_train)  # Training the model
        test_data_prediction = model.predict(X_test)  # Predicting on the test set
        accuracy = accuracy_score(Y_test, test_data_prediction)  # Calculating accuracy
        print("Accuracy score of the", model, '=', accuracy)
        # Calling the function to compare models
compare_models()

# calculating the mean of all the acuuracy of models 
cv_score = cross_val_score(LogisticRegression(max_iter = 1300), X, Y, cv =5)
print(cv_score_lr)
mean_accuracy_score = sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_score = round((mean_accuracy_score *100), 2)
print(mean_accuracy_score)
```
